```python
import time, math, random

def mergeSort(arr):
    time_complexity = []

    def merge(arr, l, m, r):
        n1 = m - l + 1
        n2 = r - m

        L = arr[l:m+1]
        R = arr[m+1:r+1]

        i = j = 0
        k = l

        while i < n1 and j < n2:
            if L[i] <= R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        while i < n1:
            arr[k] = L[i]
            i += 1
            k += 1

        while j < n2:
            arr[k] = R[j]
            j += 1
            k += 1

    def mergeSortHelper(arr, l, r, depth):
        if l < r:
            m = (l+(r-1))//2

            mergeSortHelper(arr, l, m, depth + 1)
            mergeSortHelper(arr, m+1, r, depth + 1)

            # Calculate the time complexity at this depth
            n = r - l + 1
            time_complexity.append((depth, n * math.log2(n)))  # Add time complexity at this depth

            merge(arr, l, m, r)

    time_start = time.time()
    mergeSortHelper(arr, 0, len(arr)-1, 0)
    time_end = time.time()

    return time_complexity, time_end - time_start

# Example usage
arr = [random.randint(1,10**3) for _ in range(10**5)]
time_complexity, execution_time = mergeSort(arr)
print("Time Complexity (Depth, Time):", time_complexity)
print("Execution Time:", execution_time)
```
